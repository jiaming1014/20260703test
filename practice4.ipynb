```python
import yfinance as yf

# 抓取台積電 (2330.TW) 股票資料
tsmc = yf.Ticker("2330.TW")

# 取得最近 1 個月的日線資料
hist = tsmc.history(period="1mo")
print("=== 台積電 (2330.TW) 最近 1 個月股價 ===")
print(hist.tail(10))

# 取得股票基本資訊
info = tsmc.info
print("\n=== 股票基本資訊 ===")
print(f"公司名稱: {info.get('shortName', 'N/A')}")
print(f"目前股價: {info.get('currentPrice', 'N/A')}")
print(f"52週最高: {info.get('fiftyTwoWeekHigh', 'N/A')}")
print(f"52週最低: {info.get('fiftyTwoWeekLow', 'N/A')}")
print(f"市值: {info.get('marketCap', 'N/A')}")
```

```python
import yfinance as yf
import pandas as pd

def get_tsmc_history(period: str = "1mo") -> pd.DataFrame:
    """
    取得台積電 (2330.TW) 歷史股價資料
    
    Args:
        period: 時間區間，可選 "1d"(1天), "5d"(1週), "1mo"(1月), "1y"(1年)
    
    Returns:
        DataFrame: 包含 Open, High, Low, Close, Volume 等欄位
    """
    valid_periods = ["1d", "5d", "1mo", "1y"]
    if period not in valid_periods:
        raise ValueError(f"period 必須是 {valid_periods} 其中之一")
    
    tsmc = yf.Ticker("2330.TW")
    hist = tsmc.history(period=period)
    return hist

# 測試不同時間區間
for p in ["1d", "5d", "1mo", "1y"]:
    df = get_tsmc_history(p)
    print(f"\n=== 台積電 最近 {p} ({len(df)} 筆) ===")
    print(df.tail(3))
```
